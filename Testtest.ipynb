**Prediksi Pelamar Laki-Laki dan Perempuan**

```python
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima import auto_arima
import numpy as np

# Data pelamar laki-laki dan perempuan
data_pelamar = {
    'Tahun': [2019, 2020, 2021, 2022, 2023],
    'Pelamar_Laki_Laki': [8500000, 9000000, 9500000, 10000000, 10500000],  # Contoh data
    'Pelamar_Perempuan': [6500000, 7000000, 7500000, 8000000, 8500000]  # Contoh data
}

df_pelamar = pd.DataFrame(data_pelamar)

# Memastikan tidak ada NaN dalam dataset
df_pelamar = df_pelamar.set_index('Tahun')

# Visualisasi data pelamar
plt.figure(figsize=(10, 6))
plt.plot(df_pelamar.index, df_pelamar['Pelamar_Laki_Laki'], marker='o', label='Pelamar Laki-Laki')
plt.plot(df_pelamar.index, df_pelamar['Pelamar_Perempuan'], marker='x', label='Pelamar Perempuan')
plt.title('Jumlah Pelamar Kerja di Indonesia (2019-2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pelamar')
plt.legend()
plt.grid()
plt.show()

# Menggunakan auto_arima untuk menentukan parameter yang optimal
model_laki = auto_arima(df_pelamar['Pelamar_Laki_Laki'], seasonal=False, stepwise=True, trace=True)
model_perempuan = auto_arima(df_pelamar['Pelamar_Perempuan'], seasonal=False, stepwise=True, trace=True)

# Prediksi untuk 2024 dan 2025
forecast_laki = model_laki.predict(n_periods=2)
forecast_perempuan = model_perempuan.predict(n_periods=2)

# Menampilkan hasil prediksi
years_future = np.array([2024, 2025])
prediksi_laki = pd.DataFrame(forecast_laki, index=years_future, columns=['Prediksi_Pelamar_Laki_Laki'])
prediksi_perempuan = pd.DataFrame(forecast_perempuan, index=years_future, columns=['Prediksi_Pelamar_Perempuan'])

print("Prediksi Pelamar Laki-Laki:")
print(prediksi_laki)
print("\nPrediksi Pelamar Perempuan:")
print(prediksi_perempuan)
```

**Prediksi Jumlah Lowongan Kerja dan Jumlah Pelamar**

```python
# Data lowongan kerja dan total pelamar
data_lowongan = {
    'Tahun': [2019, 2020, 2021, 2022, 2023],
    'Jumlah_Lowongan': [3000000, 3200000, 3500000, 3700000, 4000000],  # Contoh data
    'Total_Pelamar': [15000000, 16000000, 17000000, 18000000, 19000000]  # Total laki-laki dan perempuan
}

df_lowongan = pd.DataFrame(data_lowongan)

# Memastikan tidak ada NaN dalam dataset
df_lowongan = df_lowongan.set_index('Tahun')

# Visualisasi data lowongan dan total pelamar
plt.figure(figsize=(10, 6))
plt.plot(df_lowongan.index, df_lowongan['Jumlah_Lowongan'], marker='o', label='Jumlah Lowongan')
plt.plot(df_lowongan.index, df_lowongan['Total_Pelamar'], marker='x', label='Total Pelamar')
plt.title('Jumlah Lowongan Kerja dan Total Pelamar di Indonesia (2019-2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.legend()
plt.grid()
plt.show()

# Menggunakan auto_arima untuk menentukan parameter yang optimal
model_lowongan = auto_arima(df_lowongan['Jumlah_Lowongan'], seasonal=False, stepwise=True, trace=True)
model_total_pelamar = auto_arima(df_lowongan['Total_Pelamar'], seasonal=False, stepwise=True, trace=True)

# Prediksi untuk 2024 dan 2025
forecast_lowongan = model_lowongan.predict(n_periods=2)
forecast_total_pelamar = model_total_pelamar.predict(n_periods=2)

# Menampilkan hasil prediksi
prediksi_lowongan = pd.DataFrame(forecast_lowongan, index=years_future, columns=['Prediksi_Jumlah_Lowongan'])
prediksi_total_pelamar = pd.DataFrame(forecast_total_pelamar, index=years_future, columns=['Prediksi_Total_Pelamar'])

print("Prediksi Jumlah Lowongan Kerja:")
print(prediksi_lowongan)
print("\nPrediksi Total Pelamar:")
print(prediksi_total_pelamar)
```

**Prediksi Tingkat Penerimaan Pelamar**

```python
# Data untuk menghitung tingkat penerimaan
data_penerimaan = {
    'Tahun': [2019, 2020, 2021, 2022, 2023],
    'Jumlah_Pelamar': [15000000, 16000000, 17000000, 18000000, 19000000],  # Total pelamar
    'Jumlah_Diterima': [1500000, 1600000, 1800000, 2000000, 2200000],  # Jumlah yang diterima
}

df_penerimaan = pd.DataFrame(data_penerimaan)

# Menghitung tingkat penerimaan
df_penerimaan['Tingkat_Penerimaan'] = (df_penerimaan['Jumlah_Diterima'] / df_penerimaan['Jumlah_Pelamar']) * 100

# Visualisasi tingkat penerimaan
plt.figure(figsize=(10, 6))
plt.plot(df_penerimaan['Tahun'], df_penerimaan['Tingkat_Penerimaan'], marker='o', label='Tingkat Penerimaan (%)')
plt.title('Tingkat Penerimaan Pelamar Kerja di Indonesia (2019-2023)')
plt.xlabel('Tahun')
plt.ylabel('Tingkat Penerimaan (%)')
plt.legend()
plt.grid()
plt.show()

# Menggunakan auto_arima untuk menentukan parameter yang optimal
model_tingkat_penerimaan = auto_arima(df_penerimaan['Tingkat_Penerimaan'], seasonal=False, stepwise=True, trace=True)

# Prediksi untuk 2024 dan 2025
forecast_tingkat_penerimaan = model_tingkat_penerimaan.predict(n_periods=2)

# Menampilkan hasil prediksi
prediksi_tingkat_penerimaan = pd.DataFrame(forecast_tingkat_penerimaan, index=years_future, columns=['Prediksi_Tingkat_Penerimaan'])

print("Prediksi Tingkat Penerimaan Pelamar Kerja:")
print(prediksi_tingkat_penerimaan)
```
